# Patch Identification Dataset

The purpose of this script is to import and clean the above datasets. This consits of:

    - Ensuring all images are in the pixel range of [0,255]
    - Images have 3 RGB color channels, i.e. black and bhite images are convered to 3 Ch black and white images, and transparent alpha channels are dropped
    - Images are resized to a shape of 200x200 px
    
Not all classes of the WHU-RS19 and UC Merced datasets are imported. This is determined by the dictionary in `pipeline.raw_data.clean_whu_rs19` and `pipeline.raw_data.clean_uc_merced`, respectivly.

Images are then pickled and stored in training and testing files, named Xtest, Xtrain, Ytest, Ytrain to refer to ther test and train images and test and train catigorical labels, respectivly.


## UC Merced

 > This is a 21 class land use image dataset meant for research purposes. There are 100 images for each of the following classes: agricultural, airplane, baseballdiamond, beach, [etc..]. Each image measures 256x256 pixels.

## Processing WHU RS19

````python
raw_dataset = clean_utils.load_from_categorized_directory(whu_rs19_image_dir, whu_labels_to_load)
````

Because this dataset is merged with the UC Merced Aerial dataset, some of the class names are changed. In addition, some classes which are not of interest are dropped form the dataset. This behavior is defined in `pipeline.raw_data.clean_whu_rs19.py`

````python
full_sized_data = clean_utils.convert_classes(raw_dataset, clean_whu_rs19.MAP_TO_LOCAL_LABELS)
del raw_dataset # clear up some memory
whu_data = clean_utils.process_patch_for_saving(full_sized_data)
del full_sized_data
````

## Processing UC Merced

## Combind and save
